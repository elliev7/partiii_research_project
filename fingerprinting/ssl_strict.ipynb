```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
from functions import (run_multiple_avg, run_multiple_per_class_avg)
```

```python
def plot_confusion_matrix(cm, title="Confusion Matrix"):
    class_names = [str(i) for i in range(20)]
    if cm is None:
        print("No confusion matrix to display.")
        return

    cm_normalized = cm.astype('float') / cm.sum(axis=1, keepdims=True)
    cm_normalized = np.nan_to_num(cm_normalized)

    display = ConfusionMatrixDisplay(confusion_matrix=cm_normalized, display_labels=class_names)
    fig, ax = plt.subplots(figsize=(8, 6))
    display.plot(ax=ax, include_values=False, cmap='Blues')
    colorbar = display.im_.colorbar
    colorbar.set_ticks([i / 10 for i in range(11)])
    colorbar.set_ticklabels([f"{int(i * 10)}%" for i in range(11)])

    for i in range(cm_normalized.shape[0]):
        for j in range(cm_normalized.shape[1]):
            value = cm_normalized[i, j]
            ax.text(j, i, f"{value * 100:.0f}", ha='center', va='center', 
                    color='black' if value < 0.5 else 'white', fontsize=8)

    ax.set_title(title)
    plt.tight_layout()
    plt.show()
```

GLOBAL THRESHOLD -- EUCLIDEAN

```python
#euclidean thresholds 1 and 100
coverage_euclidean1, accuracy_euclidean1, train_euclidean1, cor_train_euclidean1, conf_matrix_euclidean1 = run_multiple_avg('embeddings', 'euclidean', 'distance', 100, 1, 100, 50)
print(f"Accuracy first batch: {accuracy_euclidean1[0]}")
print(f"Accuracy last batch: {accuracy_euclidean1[-1]}")
print(f"Started with {train_euclidean1[0]} training points and ended with {train_euclidean1[-1]}")
cor_new_train_euclidean1 = (cor_train_euclidean1[-1] - train_euclidean1[0]) / (train_euclidean1[-1] - train_euclidean1[0]) * 100
print(f"Percentage of correct NEW training points last batch: {cor_new_train_euclidean1:.2f}")
plot_confusion_matrix(conf_matrix_euclidean1, title="Pseudo-Labelling Confusion Matrix, Euclidean, Global, sp=1")
```
